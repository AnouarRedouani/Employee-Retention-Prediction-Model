# TASK #7: TRAIN AND EVALUATE A LOGISTIC REGRESSION CLASSIFIER

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print("Accuracy {} %".format( 100 * accuracy_score(y_pred, y_test)))
```

```python
# Testing Set Performance
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

# TASK #9: TRAIN AND EVALUATE A DEEP LEARNING MODEL

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=500, activation='relu', input_shape=(50, )))
model.add(tf.keras.layers.Dense(units=500, activation='relu'))
model.add(tf.keras.layers.Dense(units=500, activation='relu'))
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
model.summary()
```

```python
# Convertir les tableaux NumPy en DataFrames Pandas pour inspection
X_train_df = pd.DataFrame(X_train)
y_train_df = pd.DataFrame(y_train, columns=['Attrition'])

# Afficher les types de données
print("Types de données avant conversion:")
print(X_train_df.dtypes)
print(y_train_df.dtypes)

# Vérifier les premières lignes des données
print("Premières lignes des données:")
print(X_train_df.head())
print(y_train_df.head())

# Convertir les colonnes en types float si nécessaire
X_train_df = X_train_df.apply(pd.to_numeric, errors='coerce')
y_train_df = y_train_df.apply(pd.to_numeric, errors='coerce')

# Afficher les types de données après conversion
print("Types de données après conversion:")
print(X_train_df.dtypes)
print(y_train_df.dtypes)

# Remplacer les NaNs par des zéros (ou toute autre valeur appropriée)
X_train_df = X_train_df.fillna(0)
y_train_df = y_train_df.fillna(0)

# Convertir les DataFrames en tableaux NumPy
X_train = X_train_df.to_numpy()
y_train = y_train_df.to_numpy().flatten()  # Flatten pour s'assurer que y_train est un vecteur 1D

# Afficher les types de données des tableaux NumPy
print("Types de données des tableaux NumPy:")
print(X_train.dtype)
print(y_train.dtype)

# Vérifier et gérer les NaNs si présents
X_train = np.nan_to_num(X_train)
y_train = np.nan_to_num(y_train)
```

```python
model.compile(optimizer='Adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
epochs_hist=model.fit(X_train, y_train,epochs =100,batch_size=50)
```
